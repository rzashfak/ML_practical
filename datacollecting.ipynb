```python
import os
import shutil

# Define source and output paths
source_folder = "/content/dataset"  # Folder where images are stored
output_folder = "/content/categorized_images"  # Folder to save categorized images

# Create subfolders for categorized images
looking_at_camera_folder = os.path.join(output_folder, "looking_at_camera")
looking_away_folder = os.path.join(output_folder, "looking_away")
os.makedirs(looking_at_camera_folder, exist_ok=True)
os.makedirs(looking_away_folder, exist_ok=True)

# Placeholder function to determine category (update with actual logic)
def is_looking_at_camera(image_path):
    # Replace with actual logic to classify images
    # Example: Use filename, metadata, or gaze detection ML models
    return "at_camera" in image_path  # Example based on filename keyword

# Categorize images
for root, dirs, files in os.walk(source_folder):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):  # Check for image files
            image_path = os.path.join(root, file)

            # Determine category
            if is_looking_at_camera(image_path):
                target_folder = looking_at_camera_folder
            else:
                target_folder = looking_away_folder

            # Copy image to respective folder
            shutil.copy(image_path, os.path.join(target_folder, file))

print("Images categorized and copied successfully!")
```

```python
output_drive_path = "/content/drive/My Drive/categorized_images"
shutil.copytree(output_folder, output_drive_path)

print("Categorized images saved to:", output_drive_path)
```
