```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python

df =pd.read_csv("NY-House-Dataset.csv")
df
```

```python
import math
median_bedrooms = math.floor(df.BEDS.median())
median_bedrooms
```

```python
df.BEDS = df.BEDS.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['PROPERTYSQFT','BEDS', 'BATH', ]], df.PRICE)
```

```python
# prompt: reg_coef_

reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2400,3,2]])
```
